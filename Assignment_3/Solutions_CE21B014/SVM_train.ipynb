```python
import pandas as pd
from collections import Counter
import re
from nltk.corpus import stopwords
import string

def extract_words(text):
    stop_words = set(stopwords.words('english'))
    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)
    # Tokenize and convert to lowercase
    words = re.findall(r'\b\w+\b', text.lower())
    # Remove stop words
    words = [word for word in words if word not in stop_words]
    word_counts = Counter(words)
    return word_counts

def process_emails(df):
    emails_data = []
    for index, row in df.iterrows():
        content = row['Message']
        words_freq = extract_words(content)
        emails_data.append(words_freq)
    processed_df = pd.DataFrame(emails_data)
    processed_df = processed_df.fillna(0)
    return processed_df

# Read data from CSV file
data_df = pd.read_csv(r'emails_data.csv')  # Update with your CSV file path

# Process emails
processed_df = process_emails(data_df)

# Combine processed data with labels
processed_df['label'] = data_df['label']

# Save processed data to a new CSV file
processed_df.to_csv(r'processed_data.csv', index=False)
```
